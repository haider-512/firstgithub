```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")

sns.boxplot(x="survived", y="age",showmeans=True, 
            meanprops= {"marker": "+", 
                        "markersize": "12",
                        "markeredgecolor": "red"}, data=kashti)

plt.xlabel("How many survived", size=14, weight="bold"),
plt.ylabel("Age (years)", size=12, weight="bold"),
plt.title("Box plot of kitny dooby or kitny bach gaye", size =16, weight="bold")

plt.show()
```
