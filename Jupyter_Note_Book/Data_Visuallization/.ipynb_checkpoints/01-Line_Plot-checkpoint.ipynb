# Import Libraries
- Seaborn (it automatically installs the following libraries)
        Numpy
        Scipy
        Pandas
- Matplotlib

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool
```

```python
#draw a lineplot
sns.lineplot(x="sepal_length" , y="sepal_width" , data=phool)
plt.show()
```

## Adding titles

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool
#draw a lineplot
sns.lineplot(x="sepal_length" , y="sepal_width" , data=phool)
plt.title("Phoolon ka Plot")
plt.show()
```

## Adding Limits

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool
#draw a lineplot
sns.lineplot(x="sepal_length" , y="sepal_width" , data=phool)
plt.title("Phoolon ka Plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```
