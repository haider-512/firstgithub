```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True) #we have used the theme named as tick, which is already made theme

titanic = sns.load_dataset("titanic") #builtin_dataset in seaborn, we have many online datasets of seaborn
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
titanic = sns.load_dataset("titanic") #builtin_dataset in seaborn
p1=sns.countplot(x="sex", hue="class", data=titanic)
p1.set_title("Plot for counting")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
titanic = sns.load_dataset("titanic") #builtin_dataset in seaborn

g=sns.FacetGrid(titanic, row="sex", hue="alone")
g=(g.map(plt.scatter, "age", "fare").add_legend())
plt.show()
```
