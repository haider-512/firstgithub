# Import Libraries
- Seaborn automatically Install Following Libraries
*Numpy
*Pandas
*Matplotlib

```python
#import libraries

import seaborn as sns
import matplotlib.pyplot as plt

#load dataset

phool = sns.load_dataset("iris") #flowers dataset
phool

#draw a line plot
#It is always between 2 numeric values

sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.show()
```

## Adding limits

```python
#import libraries

import seaborn as sns
import matplotlib.pyplot as plt

#load dataset

phool = sns.load_dataset("iris") #flowers dataset
phool

#draw a line plot
#It is always between 2 numeric values

sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon a plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
#import libraries

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style(style=None, rc=None)

#load dataset

phool = sns.load_dataset("iris") #flowers dataset
phool

#draw a line plot
#It is always between 2 numeric values

sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon a plot")
sns.set_style("dark")

plt.show()
```

```python
#import libraries

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style(style=None, rc=None)

#load dataset

phool = sns.load_dataset("iris") #flowers dataset

#change figure
plt.figure(figsize=(8,6))
#draw a line plot
#It is always between 2 numeric values

sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon a plot")
sns.set_style("dark")

plt.show()
```
